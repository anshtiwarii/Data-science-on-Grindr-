```python
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter
import pandas as pd
import numpy as np
import matplotlib.ticker as ticker
import seaborn as sns
%matplotlib inline
```

```python
def convert2providers():
    df=pd.read_csv('finallabeleddata(modified android).csv')
    lst=[]
    for i in range(41155):
        if df['GEO_AS_0'].values[i] == 'CLARO S.A.' or  df['GEO_AS_0'].values[i] == 'TELEFÃ”NICA BRASIL' or df['GEO_AS_0'].values[i] == 'Tim Celular' or df['GEO_AS_0'].values[i] == 'Reliance Jio' or df['GEO_AS_0'].values[i] == 'Telemar Norte' or df['GEO_AS_0'].values[i]== 'Bharti Airtel' or df['GEO_AS_0'].values[i]== 'Idea Cellular' or df['GEO_AS_0'].values[i] == 'Vodafone India': #Top 8 and rest in 'OTHER' category
            lst.append(df['GEO_AS_0'].values[i]) 
        else:
            lst.append('OTHER')
    return lst
    
```

```python
def scatplot():
    lst=convert2providers()
    df['Providers'] = lst
    
    sns.scatterplot(x="mem-totalJSHeapSize", y="Providers", hue="TV_lof", data=df)
    plt.title("LOF")
    plt.xlabel("Memory")
    plt.savefig("plt1.png")
    
    sns.scatterplot(x="mem-totalJSHeapSize", y="Providers", hue="TV_hbos", data=df)
    plt.title("HBOS")
    plt.xlabel("Memory")
    plt.savefig("plt2.png")
```

```python
scatplot() #check saved pngs
```
